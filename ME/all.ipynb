```python
import numpy
import nltk
from nltk.classify import MaxentClassifier
from nltk.translate import AlignedSent, Alignment, IBMModel1

# 数据预处理
english_corpus = ["I love you", "You are beautiful", "How are you"]
chinese_corpus = ["我爱你", "你很美", "你好吗"]
english_sentences = [nltk.word_tokenize(sent) for sent in english_corpus]
chinese_sentences = [nltk.word_tokenize(sent) for sent in chinese_corpus]

# 特征提取
def extract_features(english_sent):
    features = {}
    for word in english_sent:
        features[f'has({word})'] = True
    return features

# 训练模型
training_data = []
for i in range(len(english_sentences)):
    aligned_sent = AlignedSent(english_sentences[i], chinese_sentences[i])
    training_data.append((extract_features(english_sentences[i]), aligned_sent))
classifier = MaxentClassifier.train(training_data)

# 翻译
def translate(english_sent):
    chinese_sent = []
    for word in english_sent:
        features = extract_features([word])
        chinese_word = classifier.classify(features)
        chinese_sent.append(chinese_word)
    return chinese_sent

# 测试翻译
english_sent = "I love you"
english_sent = nltk.word_tokenize(english_sent)
chinese_sent = translate(english_sent)
print(chinese_sent)
```
